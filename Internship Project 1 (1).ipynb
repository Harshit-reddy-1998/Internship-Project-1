```python
# perform upscaling to address sample imbalance
from sklearn.utils import resample
```

```python
# Here, I have done random over sampling as it increases the number of instances in the minority class.
# Separate majority and minority classes.
train_positive=df_new[df_new['label']==1]
train_negative=df_new[df_new['label']==0]

# Upsample minority class
train_negative_upsample=resample(train_negative,replace=True,n_samples=183431,random_state=18)

# Combine majority class with upsampled minority class
df1=pd.concat([train_positive,train_negative_upsample])

# Display new class counts
df1['label'].value_counts()
```

Here, the class balance has been done.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=43)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
lg.score(x_train,y_train)
```

```python
pred=lg.predict(x_test)
pred
```

```python
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
pred=dtc.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
pred=gnb.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
clf = LinearSVC(random_state=0, tol=1e-5)

clf.fit(x_train, y_train.ravel()) 
```

```python
pred=clf.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
pred=rf.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
knn=KNeighborsClassifier()  #n_neighbors=5 by default.
knn.fit(x_train,y_train)
pred=knn.predict(x_test)
print(accuracy_score(y_test,pred))
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

Thses accuracy scores can be because of overfitting. Lets check with cross validation.

# Cross-Validation

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(lg,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(dtc,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(gnb,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(clf,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(rf,x,y,cv=5)
print(score)
print(score.mean())
```

```python
score=cross_val_score(knn,x,y,cv=5)
print(score)
print(score.mean())
```

The least difference between accuracy and cv says that gaussianNB is working better.
